```python
clean = pd.read_csv('cleaned_dataset.csv')
clean.head()
```

```python
clean.isnull().sum()
```

```python
bathroom_mean = np.nanmean(clean['N_BATHROOM'])
bedroom_mean = np.nanmean(clean['N_BEDROOM'])
clean.fillna({'N_BATHROOM': bathroom_mean, 'N_BEDROOM': bedroom_mean}, inplace=True)
clean.isnull().sum()
```

CHECKING CORRELATION OF PRICE WITH ALL THE DATA COLUMNS

```python
columns_to_exclude = ['DATE_BUILD', 'AREA']  
data_subset = clean.drop(columns_to_exclude, axis=1)
sns.heatmap(data_subset.corr(), annot=True, cmap='coolwarm')
plt.show()
```

# TRAINING DATA

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection  import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr = LinearRegression()

X = clean[['INT_SQFT','Area', 'N_BEDROOM', 'N_BATHROOM',
            'N_ROOM', 'REG_FEE']]
y = clean['SALES_PRICE']

X_train, X_test, y_train, y_test =   train_test_split(X, y, test_size=0.2, random_state=1)

lr.fit(X_train,y_train)
```

```python
predictions = lr.predict(X_test)
plt.scatter(y_test,predictions)
```

```python
avg = np.average(clean['SALES_PRICE'])
print(avg)
```

TESTING AND TRAINING SCORES

```python
y_pred = lr.predict(X_test)
rmse   = np.sqrt(mean_squared_error(y_test,y_pred))
print('Training Score: ', lr.score(X_train, y_train))
print('Testing Score: ', lr.score(X_test, y_test))
print('RMSE: ',rmse)
```
